```kotlin
import kotlin.random.Random

val random = Random(0)
```

```kotlin
enum class Season {
    Winter, Spring, Summer, Fall
}

enum class Resource {
    Wood, Fiber, Fruit, Grain, Meat
}

class Job(name: String, yield: (Season) -> List<Pair<Resource, Double>>)

val consumption = listOf(
    Resource.Grain to 100.0
)

val jobs = listOf(
    Job("peasant") { season ->
        val luck = random.nextDouble(0.5, 1.5)
        if (season == Season.Fall) listOf(
            Resource.Grain to 600.0 * luck,
            Resource.Fiber to 50.0 * luck
        ) else listOf(Resource.Wood to 10.0)
    },
    Job("gatherer") { season -> listOf(Resource.Wood to 50.0) + if (season != Season.Winter) listOf(Resource.Fruit to 150.0) else listOf() },
    Job("hunter") { listOf(Resource.Meat to 150.0 * random.nextDouble(0.5, 1.5)) },
    Job("shepherd") { listOf(Resource.Meat to 100.0, Resource.Fiber to 100.0)}
)
```
