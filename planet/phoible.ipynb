```kotlin
val data = DataFrame.readCSV("C:\\Users\\Winggar\\Downloads\\phoible.csv")
```

```kotlin
data.columnNames()
```

```kotlin
val languages = data.groupBy("ISO6393")
val languagesCount = languages.groups.size().toInt()

println(
    "%affricates: ${
        languages.filter {
            it.group()
                .filter { it.strident == "+" && it.continuant == "-" }
                .isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)

println(
    "%plosives: ${
        languages.filter {
            it.group().filter {
                it.strident == "-" &&
                        it.continuant == "-" &&
                        it.nasal == "-" &&
                        it.click == "-" &&
                        it.Marginal == false &&
                        !it.Phoneme.contains("|")
            }.isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)

println(
    "%fricatives: ${
        languages.filter {
            it.group().filter {
                it.continuant == "+" &&
                        it.nasal == "-" &&
                        it.click == "-" &&
                        it.sonorant == "-" &&
                        it.Marginal == false &&
                        !it.Phoneme.contains("|")
            }.isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)

println(
    "%nasal: ${
        languages.filter {
            it.group().filter {
                it.continuant == "-" &&
                        it.nasal == "+" &&
                        it.click == "-" &&
                        it.sonorant == "+" &&
                        it.Marginal == false &&
                        !it.Phoneme.contains("|")
            }.isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)

println(
    "%semivowels: ${
        languages.filter {
            it.group().filter {
                it.continuant == "+" &&
                        it.nasal == "-" &&
                        it.click == "-" &&
                        it.sonorant == "+" &&
                        it.approximant == "+" &&
                        it.consonantal == "-" &&
                        it.Marginal == false &&
                        !it.Phoneme.contains("|")
            }.isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)

println(
    "%glides: ${
        languages.filter {
            it.group().filter {
                it.continuant == "+" &&
                        it.nasal == "-" &&
                        it.click == "-" &&
                        it.sonorant == "+" &&
                        it.approximant == "+" &&
                        it.consonantal == "+" &&
                        it.Marginal == false &&
                        !it.Phoneme.contains("|")
            }.isNotEmpty()
        }.groups.size().toInt() / languagesCount.toDouble()
    }"
)
```

```kotlin
val phonemeOccurrences = data.map { it.Phoneme }
    .toList()
    .flatMap { it.toList() }
    .distinct()
    .associateWith { symbol ->
        languages.groups.filter {
            it.filter { it.Phoneme.contains(symbol) }.isNotEmpty()
        }
    }
```

```kotlin
phonemeOccurrences
    .entries
    .map { it.key to it.value }
    .map { it.first to it.second.size().toInt() }
    .sortedByDescending { it.second }
    .map {
        "${it.first} - ${it.second} - ${
            ({
                val ratio = it.second.toDouble() / languagesCount.toDouble()
                "%.2f".format(ratio)
            })()
        }"
    }
    .forEach { println(it) }
```
